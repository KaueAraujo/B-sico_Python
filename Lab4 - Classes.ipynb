```python
# Classe
class Hangman:
    from unidecode import unidecode
    import random

	# Método Construtor
    def __init__(self):
        print('Bem vindo(a) ao jogo da FORCA!')
        print('Adivinhe a palavra abaixo antes de se enrolar...')
            
        self.palavra_chave = self.sorteiaPalavra()
        self.vidas = 6
        self.palavra_chave_limpa = []
        self.letras_descobertas = self.escondeLetras()
        self.letras_erradas = []
    
    # Método para selecionar a palavra
    def sorteiaPalavra(self):
        with open('C:/Users/Pichau/Projetos/Jogo-da-Forca\/vocabulario.txt', 'r', encoding='utf-8') as arquivo:
                conteudo = arquivo.read()
                palavras = conteudo.split('\n')
    
        return Hangman.random.choice(palavras)

	# Método para adivinhar a letra
    def letra(self):
        tentativa = input("\nEscolha uma letra: ").lower()
        return Hangman.unidecode(tentativa)

	# Método para verificar se o jogo terminou
    def gameOver(self):
        if '_' in self.letras_descobertas:
            print("Mais sorte na próxima vez! A palavra era: ", self.palavra_chave)
		
	# Método para verificar se o jogador venceu
    def playerWin(self):
        if '_' not in self.letras_descobertas:
            print("Parabéns! Você descobriu a palavra: ", self.palavra_chave)
            return True
    
	# Método para não mostrar a letra no board
    def escondeLetras(self):
        from unidecode import unidecode

        self.palavra_chave_limpa = unidecode(self.palavra_chave)
        return ['_' for letra in self.palavra_chave]

    # Método para imprimir a forca
    def displayForca(self):
    #Lista com os estágios da forca
        stages = [
        '''
            ----------
            |        |
            |        X
            |       \\|/
            |        |
            |       / \\
            |
            -
        ''','''
            ----------
            |        |
            |        0
            |       \\|/
            |        |
            |       / 
            |
            -
        ''',
        '''
            ----------
            |        |
            |        0
            |       \\|/
            |        |
            |       
            |
            -
        ''',
        '''
            ----------
            |        |
            |        0
            |       \\|/
            |        
            |       
            |
            -
        ''',
        '''
            ----------
            |        |
            |        0
            |       \\|
            |        
            |       
            |
            -
        ''',
        '''
            ----------
            |        |
            |        0
            |       
            |        
            |       
            |
            -
        ''',
        '''
            ----------
            |        |
            |        
            |       
            |        
            |       
            |
            -
        '''
        ]
        return stages[self.vidas]

	# Método para checar o status do game e imprimir o board na tela
    def play(self):

        while self.vidas > 0:
            #Preenche hífen
            if "-" in self.palavra_chave:
                index = 0
                for letra in self.palavra_chave:
                    if letra == "-":
                        self.letras_descobertas[index]="-"
                    index+=1
                
            #Print
            print(self.displayForca())
            print(" ".join(self.letras_descobertas))
            print("Vidas Restantes: ", self.vidas)
            print("Letras erradas: ", " ".join(self.letras_erradas))
            print("")

            tentativa = self.letra()

            #Verifica se a letra está na palavra
            if tentativa in self.palavra_chave_limpa:
                index = 0
                for letra in self.palavra_chave_limpa:
                    if letra == tentativa:
                        self.letras_descobertas[index]=self.palavra_chave[index]
                    index+=1
            else:
                self.vidas-=1
                self.letras_erradas.append(tentativa)

            if self.playerWin():
                break
        self.gameOver()        


```

```python
game = Hangman()
```

```python
game.play()
```
